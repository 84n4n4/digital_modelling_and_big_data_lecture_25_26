# Freefall Ode example

$\gamma=\rho A C_d$    Drag

$ma=mg-\gamma v^{2}$      Forces


$m\frac{dv}{dt}=mg-\gamma v^{2}$      Still Forces


$\frac{dv}{dt}=g-\frac{\gamma}{m} v^{2}$      Accelleration



```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.integrate import odeint


MASS = 75
G = 9.81
V0 = 0
AREA = 0.6
DRAG_COEFF = 0.5
RHO_AIR = 1

gamma = RHO_AIR*AREA*DRAG_COEFF
```

```python

def freefall(v, t, p):
    return G - p[0]/MASS * v[0]**2

times = np.arange(0, 20, 0.5)
v = odeint(freefall, t=times, y0=V0, args=tuple([[gamma]]))
```

```python

fig, ax = plt.subplots(dpi=120)
plt.plot(times, v, label="True speed", color="k", alpha=0.5)
plt.legend()
plt.xlabel("Time (Seconds)")
plt.ylabel(r"$y(t)$ (m/s)")
plt.show()
```
